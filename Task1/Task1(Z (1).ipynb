```python
#Heatmap of Total Population of a village
df_acc = pd.read_csv('531_Sangali.csv', dtype=object)
map_hooray = folium.Map(location=[16.8524, 74.5815],zoom_start = 10) 

df_acc['latitude'] = df_acc['latitude'].astype(float)
df_acc['longitude'] = df_acc['longitude'].astype(float)
df_acc['Total.Population.of.Village'] = df_acc['Total.Population.of.Village'].astype(float)
heat_df = df_acc[['Total.Population.of.Village','latitude','longitude']]
heat_df = heat_df.dropna(axis=0, subset=['Total.Population.of.Village','latitude','longitude'])
```

```python
heat_data = [[row['latitude'],row['longitude']] for index, row in heat_df.iterrows()]
#heat_data
```

```python
HeatMap(heat_data, cmap="YlGnBu").add_to(map_hooray)
map_hooray
```
